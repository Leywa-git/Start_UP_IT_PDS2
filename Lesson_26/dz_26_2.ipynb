```python
x = [10, 13, 10, 9, 9, 12, 12, 6, 7, 9, 8, 9, 11, 9, 14, 13, 9, 8, 8, 7,
    10, 10, 11, 11, 11, 12, 8, 7, 9, 10, 14, 13, 8, 8, 9, 10, 11, 11, 12, 12]
n = len(x)
```

```python
frequency = {}
for i in x:
    if i not in frequency:
        frequency[i] = 1
    else:
        frequency[i] += 1
frequency = dict(sorted(frequency.items()))
print(f"Frequency = {frequency}")
```

```python
xi = []
ni = []
freq = frequency.items()
for item in freq:
    xi.append(item[0]), ni.append(item[1])
print(f"xi = {xi}")
print(f"ni = {ni}")
```

```python
wi = []
for i in ni:
    wi.append(i/n)
print(f"wi = {wi}")
```

```python
import matplotlib.pyplot as plt
plt.title("Frequency Polygon")
plt.plot(xi, ni)
plt.show()
```

```python
plt.title("Relative Frequency Polygon")
plt.plot(xi, wi)
plt.show()
```

```python
F_x = []
cumsum = 0
for i in wi:
    cumsum +=i
    F_x.append(cumsum)
print(f"F*(x) = {F_x}")
```

```python
Mo = max(frequency, key=frequency.get)
print(f"Mode = {Mo}")
```
