```python
import plotly
import plotly.graph_objs as go
import plotly.express as px
from plotly.subplots import make_subplots
import pandas as pd
import sweetviz as sv
import ydata_profiling as yp
```

```python
cities = pd.read_excel("cities_dz_34.xlsx")
cities
```

```python
c = cities.sort_values(by="Population")
fig = make_subplots(rows=2, cols=1)
fig.update_xaxes(title_text="Population", row=1, col=1)
fig.update_yaxes(title_text ="Area (KMsqare)", row = 1, col = 1)
fig.update_xaxes(title_text="Population", row=2, col=1)
fig.update_yaxes(title_text ="Average Income (Year, $)", row = 2, col = 1)
fig.add_scatter(x=c["Population"],  y=c["Area (KMsqare)"], row=1, col=1, name = "Population to Area")
fig.add_scatter(x=c["Population"],  y=c["Average Income (Year, $)"], row=2, col=1, name = "Population to Average Income")
fig
```

```python
ci = cities.drop(columns="City")
cii = ci.drop(columns="Metro System")
cit = cii.corr()
fig = go.Figure()
fig.add_trace(go.Heatmap(x=cit.columns, y=cit.index, z=cit, colorscale='Blues'))
fig
```

```python
sweetvz = sv.analyze(cities)
sweetvz.show_notebook()
```

```python
yp.ProfileReport(cities)
```

**Pandas-profiling provides the same information as Sweetviz for a new users eye. The smallest and largest values are visible better in Sweetviz, but Pandas-profiling gives a quick info on the correlation of data in "Alerts". Also a bonus point for Pandas-profiling for "interactions".**

**For a data-analyst beginer Sweetviz is enough, due to not overcomplexing things. For a skilled data-analyst Pandas-profiling seems to fit better. I will see how it goes in the future.**

**Another bonus point to Pandas-profiling for a quick navigation at the top of report.**
