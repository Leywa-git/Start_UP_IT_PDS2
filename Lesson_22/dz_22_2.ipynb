Step 1:
---
creating 3 plain sets A for Windy, B for Rainy, C for Cold.

Creating a variable for total days in September

```python
A = set([])
B = set([])
C = set([])
days_S = 30
print(f"A = {A}")
print(f"B = {B}")
print(f"C = {C}")
```

Step 2:
--
adding the days with all 3 bad weather conditions to each set.

A ∩ B ∩ C = 1

```python
A.add(1)
B.add(1)
C.add(1)
print(f"A = {A}")
print(f"B = {B}")
print(f"C = {C}")
```

Step 3:
--
adding the days with 2 bad weather conditions to the sets accordingly the condition.

A ∩ B - A ∩ B ∩ C = 5 - 1 = 4

B ∩ C - A ∩ B ∩ C = 3 - 1 = 2

A ∩ C - A ∩ B ∩ C = 2 - 1 = 1

```python
A.add(2), A.add(3), A.add(4), A.add(5), A.add(6)
B.add(2), B.add(3), B.add(4), B.add(5), B.add(7), B.add(8)
C.add(6), C.add(7), C.add(8)
print(f"A = {A}")
print(f"B = {B}")
print(f"C = {C}")
```

Step 4:
--
adding the days with only 1 bad weather condition to the sets acordingly the condition.

A - A ∩ B - A ∩ C + A ∩ B ∩ C = 8 - 5 - 2 + 1 = 2

B - B ∩ A - B ∩ C + A ∩ B ∩ C = 12 - 5 - 3 + 1 = 5

C - C ∩ A - C ∩ B + A ∩ B ∩ C = 4 - 2 - 3 + 1 = 0

```python
A.add(9), A.add(10)
B.add(11), B.add(12), B.add(13), B.add(14), B.add(15)
print(f"A = {A}")
print(f"B = {B}")
print(f"C = {C}")
```

Step 5:
--
counting all the days in September without bad weather.

```python
A.update(B, C)
solution = days_S - len(A)
print(f"{solution} is the number of days in September without bad weather")
```
