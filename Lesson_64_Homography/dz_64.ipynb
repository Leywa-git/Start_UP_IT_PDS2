```python
import cv2

import numpy as np
import matplotlib.pyplot as plt
```

```python
img_dst = cv2.imread("chess_1.png")
pts_dst = np.array([[236, 78], [238, 220], [415, 72], [413, 229]])
```

```python
plt.figure(figsize=(10,8))
plt.imshow(img_dst)
plt.title("chess_1")
plt.axis("off")
plt.show()
```

```python
img_src = cv2.imread("chess_2.png")
pts_src = np.array([[219, 293], [388, 329], [216, 110], [456, 158]])
```

```python
plt.figure(figsize=(10,8))
plt.imshow(img_src)
plt.title("chess_2")
plt.axis("off")
plt.show()
```

```python
h, status = cv2.findHomography(pts_src, pts_dst)
h
```

```python
im_out = cv2.warpPerspective(img_src, h, (img_dst.shape[1], img_dst.shape[0]))
```

```python
plt.figure(figsize=(10,8))
plt.imshow(im_out)
plt.title("chess_homography")
plt.axis("off")
plt.show()
```
