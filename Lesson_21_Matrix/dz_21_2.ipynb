```python
A = [[4, 2, 0], [1, 3, 2], [-1, 3, 10]]
print(A)
```

```python
A[0], A[1] = A[1], A[0]
print(A)
```

```python
for i in range(len(A[0])):
    A[1][i] = A[1][i] + A[0][i] * -4
    A[2][i] = A[2][i] + A[0][i]
print(A)
```

```python
for i in range(len(A[0])):
    A[1][i] = A[1][i] // -2
    A[2][i] = A[2][i] // 6
print(A)
```

```python
A[1], A[2] = A[2], A[1]
print(A)
```

```python
for i in range(len(A[0])):
    A[2][i] = A[2][i] + A[1][i] * -5
print(A)
```
