```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

**Loading the image**

```python
def load_image(file):
    img = cv2.imread(file, cv2.IMREAD_COLOR)
    return img
```

**Showing the image**

```python
def show_image(file):
    img = load_image(file)
    img_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.axis("off")
    plt.imshow(img_RGB)
    return plt.show()
```

**Changing the contrast of the image**

```python
def adjust_contrast(img, alpha, beta):
    img_new = cv2.convertScaleAbs(img, alpha=alpha, beta=beta)
    return img_new
```

**Changing the brightness of the image**

```python
def adjust_brightness(img, brightness):
    adjusted_img = cv2.convertScaleAbs(img, beta=brightness)
    return adjusted_img
```

**Changing the saturation of the image**

```python
def adjust_saturation(img, saturation):
    hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    h, s, v = cv2.split(hsv_img)
    s = cv2.convertScaleAbs(s, alpha=saturation)
    hsv_img = cv2.merge([h, s, v])
    adjusted_img = cv2.cvtColor(hsv_img, cv2.COLOR_HSV2BGR)
    return adjusted_img
```

**Changing the size of the image**

```python
def resize_image(img, new_width, new_height):
    resized_img = cv2.resize(img, (new_width, new_height))
    return resized_img
```

**Cropping the image**

```python
def crop_image(img, x1, y1, x2, y2):
    cropped_img = img[y1:y2, x1:x2]
    return cropped_img
```

**rotating the image**

```python
def rotate_image(img, angle):
    height, width = img.shape[:2]
    rotation_matrix = cv2.getRotationMatrix2D((width/2, height/2), angle, 1)
    rotated_img = cv2.warpAffine(img, rotation_matrix, (width, height))
    return rotated_img
```
