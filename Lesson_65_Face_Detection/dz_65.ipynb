# Viola-Jones methode

```python
image = cv2.imread("titanic.jpg")
face_detector_vj = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

faces_vj = face_detector_vj.detectMultiScale(gray, 1.3, 5)

for (x, y, w, h) in faces_vj:
    cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("titanic.jpg_face_recognition")
plt.axis("off")    
plt.show()
```

```python
new_image = cv2.imread("titanic.jpg")
image_rotated = cv2.rotate(new_image, cv2.ROTATE_90_CLOCKWISE)
face_detector_vj_r = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
gray_r = cv2.cvtColor(image_rotated, cv2.COLOR_BGR2GRAY)

faces_vj_r = face_detector_vj_r.detectMultiScale(gray_r, 1.3, 5)

for (x, y, w, h) in faces_vj_r:
    cv2.rectangle(image_rotated, (x, y), (x+w, y+h), (0, 255, 0), 2)
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.imshow(cv2.cvtColor(image_rotated, cv2.COLOR_BGR2RGB))
plt.title("titanic.jpg_face_recognition_rotated_90_clockwise")
plt.axis("off")       
plt.show()
```
